```python
import pandas as pd
```

```python
    # Read data from file
    df = pd.read_csv('adult.data.csv')
    df.head(15)
```

```python
    # How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
    race_count = df.value_counts(subset=['race'])
    race_count
```

```python
    # What is the average age of men?
    average_age_men = df.loc[df['sex']=='Male', 'age'].mean()
```

```python
    # What is the percentage of people who have a Bachelor's degree?
    percentage_bachelors = df.loc[df['education']=='Bachelors', 'age'].count() / df.shape[0]
```

```python
    # What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
    # What percentage of people without advanced education make more than 50K?]
    
    # with and without `Bachelors`, `Masters`, or `Doctorate`
    higher_education = df[(df['education']=='Bachelors') | (df['education']=='Masters') | (df['education']=='Doctorate')]
    lower_education = df[(df['education']!='Bachelors') & (df['education']!='Masters') & (df['education']!='Doctorate')]
    
    # percentage with salary >50K
    higher_education_rich = higher_education[higher_education['salary'] == '>50K'].shape[0] / higher_education.shape[0]
    lower_education_rich = lower_education[lower_education['salary'] == '>50K'].shape[0]/lower_education.shape[0]
```

```python
    # What is the minimum number of hours a person works per week (hours-per-week feature)?
    min_work_hours = df['hours-per-week'].min()
   
```

```python
    # What percentage of the people who work the minimum number of hours per week have a salary of >50K?
    num_min_workers = df[(df['hours-per-week'] == 1) & (df['salary'] =='>50K')].shape[0]

    rich_percentage = (num_min_workers / df[(df['hours-per-week'] == 1)].shape[0]) *100
   
```

```python
    # What country has the highest percentage of people that earn >50K?
    highest_earning_countries = df[df['salary'] == '>50K']['native-country'].value_counts()
    highest_earning_countries_percentage = (highest_earning_countries / df["native-country"].value_counts()*100).sort_values(ascending=False)
    highest_earning_country_percentage = highest_earning_countries_percentage.max()
    highest_earning_country = highest_earning_countries_percentage.idxmax()
  
```

```python
    # Identify the most popular occupation for those who earn >50K in India.
    top_IN_occupation = df[(df['native-country'] =='India') & (df['salary'] == '>50K')]['occupation'].value_counts().idxmax()
    top_IN_occupation
                                                               
```
